# Analysis on Product 

#### Displaying the unique city in the dataset using unique fucntion iterate the city name using for loop 

```python
a=df['City'].unique()
for i in a:
    print(i)
```

#### Displaying the city along with the branches

#### Displaying the most used payment mode in dataset 

```python
df['Payment'].value_counts()
```

#### Displaying total sale by month wise 

```python
total_of_months=df.groupby('Months',as_index=False)['Total'].sum().sort_values(by="Total",ascending=False)
```

```python
plt.figure(figsize=(22,10))
sns.barplot(x="Months",y="Total",data=total_of_months)
plt.title("Total sales by months",size=25)
plt.xlabel("Months",size=20)
plt.ylabel("Total",size=20)
plt.xticks(size=15)
plt.yticks(size=15)
plt.show()
```

#### Displaying the most high quantity selling product line 

```python
high_selling_product_line=df.groupby('Product_line',as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)
```

```python
plt.figure(figsize=(22,10))
sns.barplot(x="Product_line",y="Quantity",data=high_selling_product_line)
plt.title("Top selling by productline",size=25)
plt.xlabel("Product_line",size=20)
plt.ylabel("Quantity",size=20)
plt.xticks(size=15)
plt.yticks(size=15)
plt.show()
```

#### Diplaying largest month of Cogs

```python
cogs_month=df.groupby('Months',as_index=False)['cogs'].sum().sort_values(by='cogs',ascending=False)
```

```python
plt.figure(figsize=(22,10))
sns.barplot(x="Months",y="cogs",data=cogs_month)
plt.title("Largest months of cogs",size=25)
plt.xlabel("Months",size=20)
plt.ylabel("cogs",size=20)
plt.xticks(size=15)
plt.yticks(size=15)
plt.show()
```

#### Displaying the total by product_line

```python
productline_revenue=df.groupby('Product_line',as_index=False)['Total'].sum().sort_values(by='Total',ascending=False)
```

```python
plt.figure(figsize=(22,10))
sns.barplot(x="Product_line",y="Total",data=productline_revenue)
plt.title("Total by product line",size=25)
plt.xlabel("Product_line",size=20)
plt.ylabel("Total",size=20)
plt.xticks(size=15)
plt.yticks(size=15)
plt.show()
```

#### Displaying the total by city 

```python
total_of_city=df.groupby('City',as_index=False)['Total'].sum().sort_values(by='Total',ascending=False)
```

```python
plt.figure(figsize=(22,10))
sns.barplot(x="City",y="Total",data=total_of_city)
plt.title("Total by city",size=25)
plt.xlabel("City",size=20)
plt.ylabel("Total",size=20)
plt.xticks(size=15)
plt.yticks(size=15)
plt.show()
```

#### Displaying the Largest VAT by product_line 

```python
largestVAT_productline=df.groupby('Product_line',as_index=False)['Tax_5%'].mean().round(2).sort_values(by='Tax_5%',ascending=False)
```

```python
plt.figure(figsize=(22,10))
sns.barplot(x="Product_line",y="Tax_5%",data=largestVAT_productline)
plt.title("LargestVAT_productline",size=25)
plt.xlabel("Product_line",size=20)
plt.ylabel("Tax_5%",size=20)
plt.xticks(size=15)
plt.yticks(size=15)
plt.show()
```

#### Displaying Product_category column condition based on if the total is greater than avg of total than it will be "Good" or "Bad"

```python
df['Total'].mean()
```

```python
df.loc[df['Total']>df['Total'].mean(),'Product_catogery']='Good'
df.loc[df['Total']<df['Total'].mean(),'Product_catogery']='Bad'
df.head()
```

#### Dipslaying the total sales by each branch

```python
total_by_branch=df.groupby('Branch',as_index=False)['Total'].sum().sort_values(by='Total',ascending=False)
```

```python
plt.figure(figsize=(22,10))
sns.barplot(x="Branch",y="Total",data=total_by_branch)
plt.title("Total by Branch",size=25)
plt.xlabel("Branch",size=20)
plt.ylabel("Total",size=20)
plt.xticks(size=15)
plt.yticks(size=15)
plt.show()
```

#### Displaying the product_line by Gender

```python
product_line_gender=df.groupby(['Product_line'],as_index=False)['Gender'].value_counts().sort_values(by='count',ascending=False)
```

```python
plt.figure(figsize=(22,10))
sns.barplot(x="Product_line",y='count',hue='Gender',data=product_line_gender)
plt.title("Product_line_gender",size=25)
plt.xlabel("Product_line",size=20)
plt.ylabel("Count",size=20)
plt.xticks(size=15)
plt.yticks(size=15)
plt.legend(title='Gender',fontsize=10)
plt.show()
```

#### Diplaying the average rating of each product_line 

```python
rating_product=df.groupby('Product_line',as_index=False)['Rating'].mean().round(2).sort_values(by='Rating',ascending=False)
```

```python
plt.figure(figsize=(22,10))
sns.barplot(x="Product_line",y='Rating',data=rating_product)
plt.title("Average rating of Product line",size=25)
plt.xlabel("Product_line",size=20)
plt.ylabel("Rating",size=20)
plt.xticks(size=15)
plt.yticks(size=15)
plt.show()
```

# Analysis on Sales

#### Diaplaying the sale by weekday 

```python
sale_week=df.groupby('Day',as_index=False)['Total'].sum().round().sort_values(by='Total',ascending=False)
```

#### Displaying the total sales by customer_type

```python
total_sale_cust=df.groupby('Customer_type',as_index=False)['Total'].sum().round()
```

#### Displaying the city with avg tax percent
